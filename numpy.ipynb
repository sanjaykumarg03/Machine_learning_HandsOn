```python
#Importing the numpy lib
import numpy as np
```

```python
#Creating a numpy array
np_array = np.array([3,4,2,45,551])
```

```python
np_array.shape
```

```python
a=np.array([(42,42,5,4,2),(345,3,4,5,2)],dtype=int)
print(a)
```

```python
a.shape
```

```python
#Creating a numpy array of zeros
a=np.zeros((5,3))
print(a)
```

```python
#Creating a numpy array of ones
b=np.ones((2,3))
print(b)
```

```python
#Creating a numpy array for particular values
c=np.full((5,6),10)
print(c)
```

```python
#Creating a numpy array for identity Matrix
ab=np.eye(9)
print(ab)
```

```python
#Creating a numpy array for random values
ac=np.random.random((44,33))
print(ac)
```

```python
#Creating a numpy array for random integers values
ad=np.random.randint(1,1000,(5,5))
print(ad)
```

```python
#Creating a numpy array of evenly spaced values
ae=np.linspace(10,100,50)
print(ae)
```

```python
#Creating a numpy array of evenly spaced values by step value
af=np.arange(10,100,5)
print(af)
```

```python
#Converting a list to nparray
ag=[32,434,2,4234,234]
np_array=np.asarray(ag)
print(np_array)
type(np_array)
```
