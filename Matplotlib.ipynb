```python
#Import the matplotlib 
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
a=np.linspace(1,10,100)
```

```python
b=np.sin(a)
```

```python
c=np.cos(b)
```

```python
print(a)
```

```python
print(b)
```

```python
#Plotting the sine and cosine wave graph
plt.plot(a,b)
plt.show()
```

```python
plt.plot(a,c)
plt.show()
```

```python
#adding the title for x and y labels
plt.plot(a,b , 'b*')
plt.xlabel('ANGLE')
plt.ylabel('SINE VALUE')
plt.title('SINE WAVE GRAPH')
plt.show()
```

```python
#Plotting the parabola
a=np.linspace(-10,10,100)
b=a**2
plt.plot(a,b,'g*')
plt.show()
```

```python
#multiple values in a graph
a=np.linspace(-5,5,50)
plt.plot(a,np.sin(a),'g-')
plt.plot(a,np.cos(a),'b--')
plt.show()
```

```python
#BAR GRAPH  
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
years=['1st Years','2nd Years','3rd Years','Final years']
strength=[30,20,10,39]
ax.bar(years,strength)
plt.xlabel('YEAR')
plt.ylabel('No of Students')
plt.show()
```

```python
#Pie chart
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
years=['1st Years','2nd Years','3rd Years','Final years']
strength=[30,20,10,39]
ax.pie(strength,labels=years,autopct='%1.2f%%',)
plt.show()
```

```python
#scatter plot
a=np.linspace(50,56,60)
b=np.sin(a)
c=np.cos(b)
```

```python
fig3=plt.figure()
ax=fig3.add_axes([0,0,1,1])
ax.scatter(a,b,color='g')
ax.scatter(a,c,color='r')
plt.show()
```
